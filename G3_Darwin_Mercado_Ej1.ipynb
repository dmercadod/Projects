# **Revisión Teorica**

# **Estructuras de datos**
- **Tuplas**
  : Son objetos inmutables, los cuales pueden almacenar datos, pero no se pueden modificar despues de su creación. Esta estructura se usan los ( ) para definirlas.(Thakur, 2016)

- **Diccionarios**
  : En los diccionarios podemos colectar llave y valor, donde cada valor esta directamente relacionado a una llave única. En este se debe de empezar generalmente con { } en el cual se señala primero la *Key* y luego el : *Value*.(Thakur, 2016)

- **Listas**:
  Son secuencias de objetos ordenados, donde se pueden almacenar diferente tipo de datos, los cuales se deben separar con coma y se usa [ ] para definir la lista. Las listas se deben de utilizar generalmente cuando se van a almacenar varias veces el mismo objeto sean cadenas de texto o listas de números. (Thakur, 2016)

- **Series**:
  Es una estructura unidimensional de datos que se utiliza a partir de la biblioteca Pandas, adicional a esto una serie puede tener diferente tipo de datos donde se incluyen, cadenas de texto, números, booleanos, entre otros. Son parecidos a arreglos o listas. (Thakur, 2016)

-  **DataFrame**:
  Es una estructura de datos tabular que tiene columnas y filas de manera ordenada, pudiendosé ver como un grupo de series los cuales continen indices. Generalmente los DataFrames se crean a partir de la funcion de pandas.DataFrame, siendo una de las estructura óptima para el analisis de datos, manipulación de datos y adicional para visualización de datos. (Thakur, 2016)

  **Panel**
  Es una estructura de datos para datos tridimensionales, se puede utilizar sea a partir de varios DataFrames. (Thakur, 2016)

## Creación del panel

```python
##Construya un Panel con los tres (3) dataframes cargados anteriormente: df1, df2 y df3.
panel = {"DataFrame1": df1,  ## Se desarrolla un panel con los 3 DataFrames como un diccionario, con esto definirmos los keys y los values de cada DataFrame
         "DataFrame 2" : df2,
         "DataFrame 3" : df3}
```

```python
def encabezados(panel): ## Se define la función encabezados, para luego imprimir solamente cada encabezado de las columnas de cada uno de los DataFrames
    for df_names, df in panel.items(): ## Realizamos la iteración en el Diccionario panel, donde se verifica cada item
      encabezados=df.columns ## le indicamos que utlice la función para ver todos los nombres de las columnas que tenemos.
      print("DataFrames:") ## Le ponemos nombre a cada DataFrame para mostrarlos
      print(encabezados) ## Mostramos los encabezados
```

```python
print(encabezados(panel))

##enca=encabezados(panel)
```

```python
def elementos(panel): ## Definimos la función elementos
  dim={} ## creamos una variable donde se ingresarán los datos en tipo diccionario
  for df_names,df in panel.items(): ## Realizamos la iteración en el Diccionario panel, donde se verifica cada item
      dim[df_names]=df.shape ## indicamos que utilice solamente el nombre de los DataFrames y nos de sus dimensiones
  return dim
```

```python
dim = elementos(panel) ## Variable de dimensiones
```

```python
dim ## Mostramos las dimensiones del panel
```
