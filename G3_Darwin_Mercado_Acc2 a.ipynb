- ***Desarrollo del modelo de regresión lineal***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Anexo 1 - evaluar las interrupciones del servicio de acueducto de EPM.xlsx", header = 1)
df=df[df["Municipio"] =="Medellín"] ## Definimos el municipio de interés para ajustar los datos
df.head(5)
```

```python
df2 =  df[['Motivo', 'Barrio', 'Fecha y hora esperada', 'Horas']] ## Se crea un nuevo DataFrame con las variables de interés para desarrollar el modelo
df2.head()
```

```python
# Se revisa que no hayan valores nulos en nuestro nuevo DataFrame
df2.info()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import OneHotEncoder
```

```python
#Revisamos los valores de la columna Motivo y encontramos un Cero
df2["Motivo"].unique()
```

```python
# Eliminamos el valor de cero que aparece en nuestra columa Motivo
df2 = df2[df2['Motivo'] != 0]
```

```python
#Se vuelven a revisar los datos y encontramos que ya no tenemos ese valor de cero
df2["Motivo"].unique()
```

```python
# Corregimos el tipo de dato en las horas a float ya que estaba como objeto anteriormente
df2['Horas'] = df2['Horas'].astype(float)
```

```python
# Se definen los valores de X y y para el modelo
X = df2[["Motivo","Barrio","Fecha y hora esperada"]]
y = df2["Horas"]
```

```python
# Utilizamos OneHotEncoder para codificar nuestros datos
codificador = OneHotEncoder(handle_unknown = 'ignore')
X_codificado = codificador.fit_transform(X)
codificador.get_feature_names_out()
print(X_codificado.shape)
```

```python
# Se dividen los datos en conjuntos para entrenamiento y pruebas
X_train, X_test, y_train, y_test = train_test_split(X_codificado,y,test_size=0.2)
# Se define el modelo lineal
lineal = linear_model.LinearRegression()
lineal.fit(X_train,y_train)
```

```python
# Mostramos las predicciones, los coeficientes y la precisión del modelo
ypred = lineal.predict(X_test)
print("Datos de la regresión multiple")
print()

print("valor de las pendientes de coeficientes a :")
print(lineal.coef_)

print("valor de las pendientes de coeficientes b :")
print(lineal.intercept_)

print("Presición del modelo : ")
print(lineal.score(X_train,y_train))
```

***La precisión de nuestro modelo es ideal,  por lo tanto se espera que al momento de realizar las predicciones tengamos datos correctos de cada barrio afectado por la suspensión del servicio de acueducto en el municipio de Medellín para diciembre del 2022***

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Se muestra el error cuadrático medio y el coeficiente r^2
mean_se = mean_squared_error(y_test, ypred)
r2 = r2_score(y_test, ypred)

print(f'Mean Squared Error: {mean_se}')
print(f'R-squared: {r2}')
```

```python
# Se procede a realizar una predicción para probar el modelo
X_pred = pd.DataFrame({'Motivo': ['Mantenimiento'], 'Barrio': ["SAN PABLO; LAS VIOLETAS"	], 'Fecha y hora esperada':['2022-12']})
X_pred_cod = codificador.transform(X_pred)
y_nuevo_pred = lineal.predict(X_pred_cod)

# Se rondean los valores
rounded_pred = np.round(y_nuevo_pred, 2)

print(f"Para Diciembre del año 2022 en el barrio y los motivos ingresados se esperan {rounded_pred} horas de suspensión")
```

***Se puede observar que tenemos una predicción aceptable de nuestro modelo el cual nos indica las horas de suspensión de servicio según el barrio y motivo ingresado para diciembre del 2022 que hace parte del municipio de medellín***
