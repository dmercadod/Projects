Automatice el proceso de interrupción del servicio de acueducto en función del
circuito, motivo, municipio, horas, número de instalación, impacto y nombre del

segmento de circuitos que son elegibles para la suspensión del servicio para poder
dirigirse directamente a estos circuitos.

# **Desarrollo de la acción 1**

```python
import pandas as pd
import numpy as np
```

```python
# Se carga el conjunto de datos con el cual se va a trabajar
df = pd.read_excel("Anexo 1 - evaluar las interrupciones del servicio de acueducto de EPM.xlsx",header=1)
df.head()
```

```python
# Se toman las variables especificas con las cuales se va a desarrollar nuestro modelo
auto_v = df[["Circuito", "Motivo","Municipio", "Horas", "Número de instalación", "Impacto","Nombre del contratista ","Barrio"]]
```

```python
# Se crea una copia del DataFrame Inicial
df=auto_v.copy()
```

```python
# Se verifica la información de cada variable
auto_v.info()
```

```python
# Se eliminan los datos nulos que se tienen
auto_v = auto_v.dropna()
```

```python
auto_v.info()
```

```python
auto_v.Horas = auto_v.Horas.astype(float) # Se corrige el formato de las horas a float
```

```python
auto_v["Circuito"].unique()
```

```python
auto_v["Motivo"].unique() ## Se revisan los datos de las filas y encontramos un cero, se procede a eliminar
```

```python
reemplazo = {0: "Mantenimiento"} ## Buscamos que el cero ahora quede con valor de Mantenimiento
```

```python
auto_v["Motivo"] = auto_v["Motivo"].replace(reemplazo)
```

```python
auto_v["Motivo"].unique() ## Se verifica que ya no esté el valor de cero
```

```python
auto_v["Municipio"].unique() # Se verifica los datos de municipio
```

```python
auto_v["Impacto"].unique() # Se verifica los datos del impacto
```

```python
auto_v["Horas"].unique() # Se verifican las horas
```

```python
auto_v["Número de instalación"].unique() # Se observa un valor de "-" en los datos, por lo tanto se procede a cambiar por NaN y luego eliminar esa fila
```

```python
auto_v["Número de instalación"] = auto_v["Número de instalación"].replace("-", np.nan) # Se reemplaza del número de instalación
```

```python
auto_v = auto_v.dropna() # Se elimina el NAN que se puso
auto_v.info() # Se verifica que se haya aplicado
```

```python
# Para poder ingresar los datos al modelo, primero se utiliza OneHotEncoder, para poder codificar los datos y puedan ser convertidos a variables que soporte el modelo
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
colsencoder = auto_v[["Circuito","Municipio","Impacto","Nombre del contratista ","Número de instalación","Motivo","Horas"]]
xatri = encoder.fit_transform(colsencoder)
```

```python
xatri
```

```python
# Se usará Label Encoder para la variable y que será el Barrio
from sklearn.preprocessing import LabelEncoder
labelenco = LabelEncoder()
label = labelenco.fit_transform(auto_v["Barrio"])
```

```python
y = label
X = xatri
```

```python
# Se cargan las librerías que se van a utilizar
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import svm
from sklearn.metrics import accuracy_score
from collections import Counter
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,train_size=0.2)
```

```python
# Se aplica el modelo de Maquinas de soporte vectorial
sv = svm.SVC(kernel="linear", C = 10, gamma=0.1)
sv.fit(X_train,y_train)
y_pred = sv.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)

print(f"Prediccion del conjunto de datos de prueba {y_pred}")
print(f"Prediccion del modelo SVM no lineal {accuracy}")
```

```python
# Se muestra de manera clara los segmentos que son elegibles para la suspención
print("Conteo de las predicciones por segmentos:")
# Se crea un diccionario que contará las ocurrencias de cada segmento
predicciones_por_segmentos = Counter(labelenco.inverse_transform(y_pred))
# Se itera sobre cada segmento y su conteo para mostrar la cantidad de veces que aparece cada uno
for segmentos, conteo in predicciones_por_segmentos.items():
    print(f"{conteo} repeticiones de segmentos elegibles para la suspención del servicio y poder dirigirse directamente {segmentos} ")
```
