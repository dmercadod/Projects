- **Combinación de DataFrames:**
Esta función permite combinar dos o más DataFrames en 1 solo, usando las columnas de interés como los valores claves a analizar, donde se pueden realizar operaciones y toma de datos especificos.

- **Unión de DataFrames:**
Esta función permite combinar dos DataFrames a partir de los indices, lo cual arroja los datos de interés que se busquen entre los dos DataFrames, esto se puede hacer con la función $.join()$ o $.merge()$.

- **Concatenación de DataFrames:**
La concatenación se utiliza para colocar un DataFrame uno encima de otro o también a el lado, teniendo como función principal $.concat()$

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Anexo 1 - Student Weight Status.csv", sep=",",header = 0)
df
```

**- Cree un DataFrame p1 con los datos del 0 al 1 con las columnas
'area_name' y 'county' de df.**

```python
p1= df[["AREA NAME","COUNTY"]][0:2]
```

- ***Para poder crear el DataFrame p1 y obtener los datos del 0 al 1 debimos de tomar los datos desde el cero a el 2 de las columnas AREA NAME y COUNTY, adicional a esto se puede utilizar la función .loc y esta nos da la ubicación exacta sin necesidad de ir hasta el número 2***

**- Creen un DataFrame p2 con los datos del 2 al 4 con las
columnas 'area_name' y 'county' de df.**


```python
p2= df[["AREA NAME","COUNTY"]][2:5]
p2
```

- ***Para poder realizar el DataFrame p2 y obtener los datos del 2 al 4 se utiliza el mismo metodo que en el punto anterior, empezando desde el número 2 hasta la posición 5 para obtener las posiciones desde el 2 al 4 y se filtra el DataFrame inicial por AREA NAME y COUNTY***

**- Concatene p1 y p2 y muestre los resultados.**

```python
pd.concat([p1,p2])
```

- ***Usamos la función concat para unir p1 y p2 que fueron creados anteriormente y poder visualizar todos los datos combinados***

**- Concatene p1 y p2 identificando las asignaciones por las llaves
‘p1’ y ‘p2’ y muestre los resultados y asígnelos a la variable
concatened. Comente las diferencias con el DataFrame anterior
y mencione desde su punto de vista las posibles utilidades de
esta herramienta.**


```python
concatenated = pd.concat([p1,p2], keys={"p1","p2"})
print(concatenated)
```

- ***En este nuevo DataFrame lo que tenemos es que hay unas llaves especificas para cada valor de p1 y p2, a diferencia del pasado donde los datos erán simplemente un DataFrame común.***
- ***Esta herramienta sería muy útil para diferencia ciertos datos que sean parecidos en un DataFrame que tenga datos por ejemplo de notas de colegio que sean parecidas y con esto podemos diferencia a cual estudiante pertenece cada nota y no tener complicaciones***

**- Desde el DataFrame concatened rescate los datos de p1.**

```python
concatenated.loc["p1"]
```

- ***Se pudo rescatar los datos de p1 a través de la función .loc que nos arroja la posición y valores exactos de p1***

# **UNIONES (JOINS):**

**- Defina un DataFrame de búsqueda llamado grade_lookup, que
tenga las columnas ‘grade_level y ‘level’ que asigne niveles a
cada uno de los grados como sigue: 1 a 'ELEMENTARY', 2 a
'MIDDLE/HIGH', 3 a 'MISC'.**

```python
datos= {"GRADE LEVEL" : ["ELEMENTARY", "MIDDLE/HIGH", "MISC"], "level" :[1,2,3] }
grade_lookup=pd.DataFrame(datos)
grade_lookup
```

***- Se definieron los keys values de GRADE LEVEL Y level que son las variables de interés con las que vamos a trabajar, luego se definieron como un DataFrame, para poder trabajar sobre los datos mas adelante***

**- Por otro lado, tome las primeras cinco filas de la columna de
datos en la columna GRADE e imprímalas (Téngalas en cuenta
pues las utilizaremos para estudiar los tipos de uniones).**

```python
prueba = pd.DataFrame(df["GRADE LEVEL"].loc[0:4])
```

***- Se utiliza la función .loc para tomar los variables de las primeras 5 columnas y se muestran***

**- Describa los inner joins y realice un ejemplo con las primeras 5
filas de la columna grade_level y grade_lookup, muestre sus
resultados y explíquelos.**

- **Inner joins**: Se utilizan para encontrar un valor en especifico que tienen las dos tablas o DataFrames propuestos y se debe de realizar con la función .join() especificando que es será how = "inner".

```python
inn_join = prueba.join(grade_lookup.set_index(["GRADE LEVEL"]), on="GRADE LEVEL", how="inner")
inn_join
```

***Como se puede observar en la tabla obtenida, tenemos los datos que tienen en común las dos tablas, teniendo como función .join() se defininen las tablas que se van a usar y el metodo que en este caso es how = "inner"***

**- Describa los left outer joins y realice un ejemplo con las
primeras 5 filas de la columna grade_level y grade_lookup,
muestre sus resultados y explíquelos.**

- **Los left outer joins nos permiten determinar los valores en especifico de esa tabla y verificar con otra que referente hay en ambas partes, teniendo como resultado una tabla con todas las variables de la tabla de estudio.**

```python
inn_joinleft = prueba.join(grade_lookup.set_index(["GRADE LEVEL"]), on="GRADE LEVEL", how="left")
inn_joinleft
```

***Se utilizo la misma función .join() y se tomo como how = "left" ya que era un left join, donde analizamos los datos completos de la tabla GRADE LEVEL y la relación de grade_lookup***

**- Describa los full outer joins y realice un ejemplo con las
primeras 5 filas de la columna grade_level y grade_lookup,
muestre sus resultados y explíquelos.**

- **En los full outer joins se utiliza los datos de ambas tablas en común, sin tener en cuenta sus relaciones entre si, solamente los datos individuales de cada uno, teniendo como resultado el total de valores que contiene cada tabla en una sola.**

```python
inn_joinouter = prueba.join(grade_lookup.set_index(["GRADE LEVEL"]), on="GRADE LEVEL", how="outer")
inn_joinouter[["GRADE LEVEL", "level"]]
```

***En esta función se volvío a utilizar .join() junto con how = "outer" que nos indica que se hará un full outer join, arrojandonos los datos de ambas tablas sin necesidad de que haya una relación entre si.***

- Encontrar la suma del número de estudiantes obesos en cada
uno de los grados mediante el uso de la función groupby de
