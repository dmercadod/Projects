# **Revisi√≥n te√≥rica**

## - Graficos de dispersi√≥n:

Este tipo de gr√°fico representan dos variables como puntos en un plano y se puede ver su relaci√≥n entre ellas. Adicional a esto son gr√°ficos √≥ptimos para representar grupos y conglomerados. Se puede utilizar la funci√≥n .scatter de la librer√≠a matplotlib.pyplot para obtener un gr√°fico de dispersi√≥n.(Thakur, 2016)

## - Histogramas

Los histogramas son gr√°ficos que se utilizan para representar de manera efectiva la distribuci√≥n de una variable. Generalmente se toma el rango de valores y se dividen en intervalos de tama√±o igual, y la elevaci√≥n de cada intervalo corresponde a la frecuencia de valores dentro de ese intervalo. (Thakur, 2016)

## - Gr√°fico de barras

El gr√°fico de barras es utilizado para mostrar datos agrupados a partir de barras rectanguales, que pueden ser verticales o horizontales, y la elevaci√≥n o longitud de estas corresponden a sus valores.(Thakur, 2016)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Gr√°fico 1

```python
x = [13,14,15,16,17,18]
y = [16,24,45,57,75,102]
```

```python
def fx(x): ##funcion lineal
    return 17*x - 210.3333

def gx(x): ## funcion exponencial
    return 0.1452 * np.exp (0.369*x)

def hx(x): ## funcion logaritmica
    return 259.49*np.log(x) - 656.46

##Dominio solicitado de [13,18]
dominio = np.linspace(13,18,100)
lineal = fx(dominio)
exponencial = gx(dominio)
logaritmica=hx(dominio)

##Graf√≠co de puntos de dispersi√≥n
plt.scatter(x, y,label = "Puntos de dispersion", color='green')

## Gr√°fico de funci√≥n l√≠neal
plt.plot(dominio, lineal, label='f(x) = 17x - 210.3333', color='yellow')

## Gr√°fico de funci√≥n exponencial
plt.plot(dominio,exponencial, label = "g(x) = 0.1452*exp^(0.369*x)",color = "red" )

## Gr√°fico de funci√≥n logaritmica
plt.scatter(dominio,logaritmica, label = "h(x) 259.49ln(x)-656.46", color = "pink")

## Agregamos la legenda en el gr√°fico
plt.legend()

## Se da el nombre al eje x y y
plt.xlabel("x")
plt.ylabel("y")

## Se agrega el titulo a la gr√°fica
plt.title("Grafico 1")

## Se muesta el grafico
plt.show()
```

# Gr√°fico 2

```python
## Generar 4 listas aleatorias de 700 muestras (revise la documentaci√≥n de numpy.random):
media=5
desve= 0.1
x1= np.random.normal(loc=media,scale=desve, size = 700)
media2=7
desvie2=0.1
y1= np.random.normal(loc=media2,scale=desvie2, size = 700)
x2=np.random.uniform(low=(min(x1)),high=(max(x1)), size= 700)
y2=np.random.uniform(low=(min(y1)),high=(max(y1)), size = 700)
```

```python
##Graficar en un mismo plot nubes de dispersi√≥n de los pares de puntos que se forman con (x1, y1) y (x2, y2)
plt.scatter(x1,y1, label = "Distribuci√≥n normal de (x1,y1)", s= 30,color ="red")
plt.scatter(x2,y2, label = "Distribuci√≥n uniformen (x2,y2)", s=30 , color = "yellow")
plt.title("Gr√°fico 2")
plt.legend()
plt.show()
```

```python
##La nube de puntos (x2, y2) debe visualizarse sobre la nube de (x1, y1).
plt.scatter(x1,y1,label = "x1,y1", color = "yellow")
plt.scatter(x2,y2,label = "x2,y2", color ="red")
plt.title("Gr√°fico 2")
plt.legend()
plt.show()
```
