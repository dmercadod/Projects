-  **¿Qué es REST web service y cómo se utilizan?**
  - *Es una interface que permite conectar varios sistemas en el protocolo HTTP, tales como GET, POST, PUT y DELETE y su representación como URLs. (Madhavan, 2015)*
- **¿Qué es una RESTful API?**
  - *Un RESTful API es un concepto la cual es una interfaz que permite que las aplicaciones compartan y tengan comunicación , las cuales son desarrolladas en base a recursos que son representados como un URLs.(Madhavan, 2015)*
-**¿Qué es JSON?**
  - *JSON significa JavaScript Object Notation, el cual es un formato para intercambio de datos que es ligero y puede ser leído por las personas, esta información se presenta en forma ordenada y con jerarquia.(Madhavan, 2015)*

```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

***Se crea la variable SpaceX con la función requests.get que nos permite obtener los datos de la URLs propuesta y así poder cargarlos***

```python
SpaceX = requests.get("https://api.spacexdata.com/v4/ships")
SpaceX
```

***Podemos observar en la salida que la response es 200, lo cual significa que fue cargada con éxito***

**- Cargue el json de la respuesta del request anterior como un
DataFrame de pandas (ayuda: función pd.DataFrame.from_dict).**

***Inicialmente usamos la función .json() para definir los datos de la variable SpaceX como JSON***

```python
SpaceX.json()
```

***Luego se desarrolla la creación de la variable naves, la cual contiene los datos de las naves y está ahora presentado como un DataFrame***

```python
naves = pd.DataFrame.from_dict(SpaceX.json())
naves.head(29)
```
