```python
Familia = {"Gabriela":( 1, 12, 77, False), ## Se crea un diccionario con tuplas de 4 donde se agregan los datos de Edad, Peso, Estatura y un dato bolean
         "Dilia": (50, 65, 169, False),
         "Golla":( 67, 85, 170, False), 
         "Darwin":(26, 95, 183, True),
         "Luis":( 29, 95, 188, True), 
         "Mayra":( 27, 75, 160, False), 
         "Raul":( 53, 80, 170, False),
         "Daniel":( 24, 182, 85, False)}
Familia # Llamamos el diccionario creado
```

## Punto b.

```python
limite = 1 ## ponemos un limite a la iteración que vamos a crear
Edad = [] ## creación de variables donde vamos a almacenar los datos de la familia
Peso = []## creación de variables donde vamos a almacenar los datos de la familia
Estatura = []## creación de variables donde vamos a almacenar los datos de la familia
Estudia = []## creación de variables donde vamos a almacenar los datos de la familia

for personas, datos in Familia.items(): ## Ciclo for en el cual se usa el DICT familia para iterar sobre los datos con .items()
  edad,peso,estatura,estudia = datos ## agregamos un orden
  for i in range (limite): ## le damos un limite a la iteración
    Edad.append(edad) ## ingresar la edad
    Peso.append(peso) ## ingresar el peso
    Estatura.append(estatura) ## ingresa la estatura
    Estudia.append(estudia)## ingresa el dato de si estudia o no


print("Edad :" , Edad) ##imprimimos los datos
print("Peso :" , Peso)##imprimimos los datos
print("Estatura :" , Estatura)##imprimimos los datos
print("Estudia :" , Estudia)##imprimimos los datos
```
