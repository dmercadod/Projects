***Inicialmente llamamos las librerias y funciones de interés, adicional a esto se cargaron los datos de la api que contiene los datos de las naves de SpaceX y lo convertimos en un DataFrame***

```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

***Se realiza el filtro sobre los datos de la nave y se realiza un nuevo df con las variables de interés***

```python
df = naves[["type", "mass_kg", "year_built","home_port", "name"]]
df
```

# **Revisando datos faltantes:**

**- Determine cuantos datos faltantes hay en la variable mass_kg del mediante value_counts()**


***Utilizamos la función value_counts() sobre los datos de la columna mass_kg e indicamos que no excluya los valores nulos***

```python
df["mass_kg"].value_counts(dropna=False)
```

- ***En la columna mass_kg tenemos que hay 8 valores faltantes***

-**Crea un DataFrame df1 removiendo las filas que tienen algunos
datos faltantes. Por otro lado crea un DataFrame df2 removiendo
las filas que tienen todos los datos faltantes. Compáralos y
comenta los resultados.**

- ***Se crea un nuevo df1 en el cual se van a eliminar las filas que tienen datos faltantes* con la función .dropna()**

```python
df1= df.dropna() ## Creamos un Dataframe eliminando las filas que tienen un dato faltante
```

```python
df1
```

- ***Se realiza una nueva variable df2 en la cual vamos a definir que se eliminen todas las filas con datos faltantes, para esto utilizamos la función dropna() junto con how = "all" que nos permite indicar que sea toda la fila con datos faltantes***

```python
df2 = df.dropna(how="all") ## Eliminamos todas las filas que tengan un dato faltante
```

```python
df2
```

- ***Se muestra la longitud de df1 y df2 para poder analizarlos correctamente***

```python
print("DataFrame df1:")
print(len(df1))


print("\nDataFrame df2:")
print(len(df2))
```

- ***Se pudo observar que la diferencia entre df1 y df2 es que en df1 se eliminaron las filas que tuvieran un dato faltante y pudimos observar que las dimensiones cambiaron donde tenemos 21 filas para df1 y 29 filas para df2, esto nos dice que en df2 no todas las filas tenían datos nulos, por lo tanto este DataFrame quedó igual.***

- **En el DataFrame df completa los datos faltantes por los siguientes
tres métodos (muestra y comenta los resultados):**

  - Completar datos faltantes con el número 0.

  - Completar por método de propagación hacia adelante.

  - Completar los valores faltantes con el promedio de la
columna.

**-Completar datos faltantes con el número 0.**

- ***Se completaron los datos faltanes inicialemente con ceros utilizando la función fillna que nos permite completar los datos dependiendo un metodo especifico.***

```python
df.fillna(0)
```

**- Completar por método de propagación hacia adelante.**

- **Para poder usar el metodo de propagación hacía adelante usamos el metodo "ffill", aunque también "pad" nos pudo haber funcionado para completar los datos faltantes por el método de propagación hacía adelante.**

```python
df.fillna(method="ffill")
```

**- Completar los valores faltantes con el promedio de la columna.**

***Para completar los valores faltantes con el promedio de la columna, se utiliza la función .fillna y se agrega el promedio con df.mean, indicamos que solamente utilice los datos de las columnas numericas solamente***

```python
df.fillna(df.mean(numeric_only=True))
```

#**Operaciones de strings:**

**- De la columna home_port extraiga la primera palabra de cada
dato y muestre los resultados.**

- ***Definimos un nuevo df con los datos de home_port y se utiliza la función str.extract para extraer el string y damos como limite que sea un valor alfanúmerico y llegue hasta donde va el espacio que es donde va nuestra primera palabra de cada dato con $'(\w+)'$***

```python
df_new= df["home_port"].str.extract('(\w+)')
df_new
```

- ***Se puede observar que la función desarrollada nos arroja el dato de la primera palabra de cada una de las filas***
